```python
# seperate out the categorical and numerical features

numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = data.select_dtypes(include=[object]).columns.tolist()

print (numerical_features)
print (categorical_features)
```

```python
# Lets add 2 new columns to the dataset, one for the average score and another for the total score of each student

data['average score'] = data[['math score', 'reading score', 'writing score']].mean(axis=1)
data['total score'] = data[['math score', 'reading score', 'writing score']].sum(axis=1)
```

```python
# lets find out number of students whho got full marks in reading, writing and math

print ( f"Number of student with full marks in reading are {data[data['reading score']==100].shape[0]}")
print ( f"Number of student with full marks in writing are {data[data['writing score']==100].shape[0]}")
print ( f"Number of student with full marks in math are {data[data['math score']==100].shape[0]}")
```

```python
# create a graph to show the distribution of reading, writing and math scores

sns.set_style('whitegrid')
sns.histplot(data=data, x='reading score', kde=False, color='green', bins=20, hue='gender', alpha=0.5)
```
